### Extract Article Category

```python
# extract Article Category
def extract_category(row):
    return row[0].get("category")


df0["category"] = df0["instances"].apply(extract_category)
df1["category"] = df1["instances"].apply(extract_category)
df2["category"] = df2["instances"].apply(extract_category)
df3["category"] = df3["instances"].apply(extract_category)
df4["category"] = df4["instances"].apply(extract_category)
df5["category"] = df5["instances"].apply(extract_category)
df6["category"] = df6["instances"].apply(extract_category)
df0.head()
```

### Extract Year and Month from "Published At" Column

```python
# view the first instance of PublishedAt
df0["publishedAt"].iloc[0]
```

```python
# Extract year and month from publishedAt column

def extract_year_month_efficient(df):
    # Convert to datetime with error handling
    dates = pd.to_datetime(df['publishedAt'], errors='coerce')
    
    # Extract year and month vectorized
    df['year'] = dates.dt.year.astype('Int64')  
    df['month'] = dates.dt.month.astype('Int64')  
    
    return df


for i, df in enumerate([df0, df1, df2, df3, df4, df5, df6]):
    extract_year_month_efficient(df)
    
# Check the results
df0.head()
```

### Clean Title Columns

```python
# condenses titles into one column
def get_title(row):
    if row["en-title"] is None:
        return row["title"]
    else:
        return row["en-title"]

df0["new_title"] = df0.apply(get_title, axis=1)
df1["new_title"] = df1.apply(get_title, axis=1)
df2["new_title"] = df2.apply(get_title, axis=1)
df3["new_title"] = df3.apply(get_title, axis=1)
df4["new_title"] = df4.apply(get_title, axis=1)
df5["new_title"] = df5.apply(get_title, axis=1)
df6["new_title"] = df6.apply(get_title, axis=1)
```

```python
df6[df6["en-title"].isna()].head()
```

```python
for df in [df0, df1, df2, df3, df4, df5, df6]:
    df.drop('en-title', axis=1, inplace = True)
    df.drop('title', axis=1, inplace = True)
```
